Download database file.

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

df_desire = pd.read_sql_query(QUERY,conn)
df_desire
```

```python
df_top5data = df_top5data.truncate(after=4)
df_top5data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()

df_top5data.plot(kind='pie',y='Respondent',labels=df_top5data['DatabaseDesireNextYear'], colors=['r', 'g', 'b', 'c','m'], autopct='%.2f');
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
df = df[df['CodeRevHrs'].notna()]
df.head(10)
```

```python
df_30_35=df[df['Age']<36]
df_30_35=df_30_35[df_30_35['Age']>29]
df_30_35.head(10)
```

```python
df_median=df_30_35.groupby("Age").median()
df_median
```

```python
import numpy as np
from matplotlib import rc

sns.set_style(style='white') 

rc('font', weight='bold')
 
bars1 = df_median['WorkWeekHrs']
bars2 = df_median['CodeRevHrs']
 
# Heights of bars1 + bars2
bars = np.add(bars1, bars2).tolist()
 
# x-axis position of bars
r = [0,1,2,3,4,5]
 
# define x axis labels
names = df_median.index
barWidth = 1
 
# blue bars
plt.bar(r, bars1, color='#0000FF', edgecolor='white', width=barWidth)
# cyan bars above blue bars
plt.bar(r, bars2, bottom=bars1, color='#00FFFF', edgecolor='white', width=barWidth)
 
# X and Y axis
plt.xticks(r, names, fontweight='bold')
plt.xlabel("Age")
plt.ylabel("Hours")
 
# Show graphic
plt.show()
```
